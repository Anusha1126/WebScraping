```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the cleaned CSV file

df = pd.read_csv('cleaned_products.csv')

# Calculate the average price of the products
average_price = df['Price'].mean()
print(f"The average price of the products is ${average_price:.2f}")

# Determine the price range category with the most products
most_common_price_range = df['Price Range'].mode()[0]
print(f"The most common price range category is '{most_common_price_range}'")

# Generate a bar plot showing the number of products in each price range category
plt.figure(figsize=(10, 6))
sns.countplot(x='Price Range', data=df, palette='viridis')
plt.title('Number of Products in Each Price Range Category')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.show()
```
